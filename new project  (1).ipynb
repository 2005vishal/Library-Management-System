```python
def Amount():
    no_of_day = 0
    while 1: 
        start = input('Enter Issue Date: ')
        end = input('Enter Deposite Date: ')
        if datetime.datetime.strptime(end, '%d/%m/%Y') > datetime.datetime.strptime(start, '%d/%m/%Y'):
            range_date = list(pd.date_range(start,end))
            for i in range(len(range_date)):
                no_of_day+=1
            break
        else: print('Invalid Date')
    if no_of_day <= 15:
        return no_of_day*5
    else:
        day = no_of_day - 15
        return day*10 + 75
```

```python
# this code is reposible for payment calculatition
def deposite_fare_calculation(start,end):
    range_date = list(pd.date_range(start,end))
    no_of_day = 0
    for i in range(len(range_date)):
        no_of_day+=1
    if no_of_day <= 15:
        return no_of_day*5
    else:
        day = no_of_day - 15
        return day*10 + 75
```
